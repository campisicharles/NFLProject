```python
import numpy as np
import pandas as pd
from sklearn.externals import joblib

model1=joblib.load('John_Madden.joblib') 
model2=joblib.load('Shotgun.joblib') 
model3=joblib.load('Pro_Set.joblib') 
model4=joblib.load('Points.joblib') 
game_data = pd.read_csv('Week8Matchups.csv')
game_data.head()
```

```python
week7X = game_data[['HomeLine', 'PredictTotal', 'HOME_OFFPass_DVOA', 'HOME_OFFRush_DVOA', 'HOME_DEFPassDVOA', 'HOME_DEFRushDVOA', 'HOME_STDVOA', 'VISIT_OFFPass_DVOA', 'VISIT_OFFRush_DVOA', 'VISIT_DEFPassDVOA', 'VISIT_DEFRushDVOA', 'VISIT_STDVOA']]
diffGuess = model1.predict(week7X)
pointsGuess= model4.predict(week7X)
betGuess = model2.predict(week7X)
classedGuess = model3.predict(week7X)

diffGuess
```

```python
diffGuesses = pd.DataFrame({"Madden Differential": diffGuess, "Over/Under": pointsGuess, "Home Cover?": betGuess, "Class Guess":classedGuess})[["Madden Differential", "Over/Under", "Home Cover?", "Class Guess"]]
diffGuesses
```

```python
nice = pd.concat([game_data, diffGuesses], axis=1, sort=False)
cleaned = nice[["Home","Away","HomeLine","PredictTotal","Madden Differential", "Over/Under", "Home Cover?", "Class Guess"]]
games = cleaned.dropna()
games.insert(8, "Year", "2019")
games.insert(9, "Week", "8")
games.to_csv(path_or_buf="Week8Predictions.csv", index=False)
```
