```python
import numpy as np
import pandas as pd
```

```python
game_data = pd.read_csv('Game_Data2.csv')
game_data.head()
```

```python
X = game_data[['HomeLine', 'HOME_OffPassDVOA', 'HOME_OffRushDVOA', 'HOME_DefPassDVOA', 'HOME_DefRushDVOA', 'HOME_STDVOA', 'VISIT_OffPassDVOA', 'VISIT_OffRushDVOA', 'VISIT_DefPassDVOA', 'VISIT_DefRushDVOA', 'VISIT_STDVOA' ]]
y = game_data['Covered']
print(X.shape, y.shape)
target_names = ["Bet Away", "Bet Home"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=29)
```

```python
from sklearn.svm import SVC 
from sklearn.model_selection import GridSearchCV
model = SVC(kernel='linear')
param_grid = {'C': [ 0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000],
              'gamma': [ 0.00001,0.0001, 0.001, 0.01, 0.1, 1, 10]}
grid = GridSearchCV(model, param_grid, verbose=3, scoring = 'f1')
```

```python
grid.fit(X_train, y_train)
```

```python
from sklearn.externals import joblib
joblib.dump(grid, 'Shotgun2.joblib') 
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
thursX = pd.read_csv('Week7clean.csv')
thurs = grid.predict(thursX)
thurs
```

```python
from sklearn.metrics import classification_report
predictions = grid.predict(X_test)
print(classification_report(y_test, predictions,
                            target_names=target_names))
```

```python
guesses = pd.DataFrame({"Predicted": predictions, "Actual": y_test})[["Predicted", "Actual"]]
guesses
```
